# 6-scale data

```python
data = data.map(lambda x , y :( x/ 255, y))
data.as_numpy_iterator().next()
```

# 7-split data

```python
train_size=int(len(data)*0.7)
evaluation_size=int(len(data)*0.2)
test_size=int(len(data)*0.1)
train_size, evaluation_size,test_size
```

```python
train = data.take(train_size)
evaluation=data.skip(train_size).take(evaluation_size) 
test= data.skip(train_size+evaluation_size).take(test_size) 
```

# 8-Model part 
## A-build the model

```python
model =  Sequential()
```

Note: CNN part

```python
model.add(Conv2D(16 , (3,3) , activation='relu' , input_shape=(256,256,3)))

model.add(MaxPooling2D())
model.add(Conv2D(32 , (3,3) , activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16 , (3,3) , 1 , activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256 , activation='relu'))
model.add(Dense(1 , activation='sigmoid'))
```

```python
model.compile('adam' , loss= tf.losses.BinaryCrossentropy() , metrics=['accuracy'])
```

```python
model.summary() 
```

## B-train the model 

```python
log_directory ='logs' 
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_directory)
```

```python
history = model.fit(train , epochs=20 , validation_data=evaluation , callbacks=[tensorboard_callback])
```

# 9- model performance

```python
fig = plt.figure()
plt.plot(history.history['loss'] , label='loss' , color='green')#show loss results in red and name it loss
plt.plot(history.history['val_loss'] , label='val_loss' , color='red')
fig.suptitle('Loss') #figure title
plt.legend(loc = 'upper left') 
plt.show() 
```

```python
fig = plt.figure()
plt.plot(history.history['accuracy'] , label='accuracy' , color='green')#show loss results in red and name it loss
plt.plot(history.history['val_accuracy'] , label='val_accuracy' , color='red')
fig.suptitle('accuracy') #figure title
plt.legend(loc = 'upper left') 
plt.show() 
```

# 10-evaluation of data

```python
pre = Precision()
rec = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    x , y = batch
    yhat = model.predict(x)
    pre.update_state(y , yhat)
    rec.update_state(y , yhat)
    acc.update_state(y , yhat)
```

```python
print(pre.result() , rec.result() , acc.result())
```
